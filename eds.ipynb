```python
import pandas as pd
from plotnine import ggplot, aes, geom_line, labs, theme_minimal, theme, scale_color_manual, geom_point, geom_text

# Read the datasets
demographics = pd.read_csv('data/2023-09-05/demographics.csv')
wages = pd.read_csv('data/2023-09-05/wages.csv')
states = pd.read_csv('data/2023-09-05/states.csv')

```

```python


# Filter data for all workers
all_workers = demographics[demographics['facet'] == 'all wage and salary workers']

plot1 = (ggplot(all_workers, aes(x='year', y='p_members')) +
         geom_line(color="blue", size=1.5) +
         geom_point(color="blue", size=2) +
         labs(title="Union Membership Over Time (All Workers)", 
              x="Year", 
              y="Percentage of Union Members (%)") +
         theme_minimal() +
         theme(figure_size=(10, 6)))

# Display the plot in a notebook
plot1.show()
```

```python
# Union Membership by Sector
sectors = ['construction', 'manufacturing', 'services', 
           'transportation, communication, and utilities', 
           'finance, insurance, and real estate']

sector_data = demographics[demographics['facet'].isin(sectors)]

plot2 = (ggplot(sector_data, aes(x='year', y='p_members', color='facet')) +
         geom_line(size=1.2) +
         labs(title="Union Membership by Sector Over Time", 
              x="Year", 
              y="Percentage of Union Members (%)", 
              color="Sector") +
         scale_color_manual(values=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']) +
         theme_minimal() +
         theme(figure_size=(10, 6)))


plot2.show()
```

```python
# Union Membership by Demographics
demographics_groups = ['demographics: male', 'demographics: female', 
                       'demographics: white male', 'demographics: black male', 
                       'demographics: hispanic male']

demographic_data = demographics[demographics['facet'].isin(demographics_groups)]

plot3 = (ggplot(demographic_data, aes(x='year', y='p_members', color='facet')) +
         geom_line(size=1.2) +
         labs(title="Union Membership by Demographics Over Time", 
              x="Year", 
              y="Percentage of Union Members (%)", 
              color="Demographic Group") +
         scale_color_manual(values=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']) +
         theme_minimal() +
         theme(figure_size=(10, 6)))
plot3.show()
```

```python
# Wage Difference Between Union and Non-Union Workers
wage_all_workers = wages[wages['facet'] == 'all wage and salary workers']

plot4 = (ggplot(wage_all_workers) +
         geom_line(aes(x='year', y='union_wage', color='"Union Wage"'), size=1.5) +
         geom_line(aes(x='year', y='nonunion_wage', color='"Non-Union Wage"'), size=1.5) +
         labs(title="Mean Hourly Wage Over Time (Union vs Non-Union)", 
              x="Year", 
              y="Mean Hourly Wage ($)", 
              color="Wage Type") +
         scale_color_manual(values=['#1f77b4', '#ff7f0e']) +
         theme_minimal() +
         theme(figure_size=(10, 6)))


plot4.show()
```

```python
# Union Wage Premium Over Time
plot5 = (ggplot(wage_all_workers, aes(x='year', y='union_wage_premium_adjusted')) +
         geom_line(color="green", size=1.5) +
         geom_point(color="green", size=2) +
         labs(title="Union Wage Premium Over Time", 
              x="Year", 
              y="Adjusted Wage Premium (%)") +
         theme_minimal() +
         theme(figure_size=(10, 6)))

plot5.show()
```

```python
# State-Level Union Membership Over Time
states_to_plot = ['New York', 'California', 'Texas', 'Michigan', 'South Carolina']
state_data_filtered = states[(states['state'].isin(states_to_plot)) & (states['sector'] == 'All')]

plot6 = (ggplot(state_data_filtered, aes(x='year', y='p_members', color='state')) +
         geom_line(size=1.2) +
         labs(title="Union Membership by State Over Time", 
              x="Year", 
              y="Percentage of Union Members (%)", 
              color="State") +
         scale_color_manual(values=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']) +
         theme_minimal() +
         theme(figure_size=(10, 6)))


plot6.show()
```
