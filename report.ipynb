```python
from plotnine import *
import pandas as pd

# Create a sample dataset
df = pd.DataFrame({
    'x': range(1, 11),
    'y': [3, 5, 7, 8, 4, 2, 6, 7, 8, 10],
    'category': ['A']*5 + ['B']*5
})

# Create a scatter plot with plotnine
plot = (ggplot(df, aes(x='x', y='y', color='category'))
        + geom_point()
        + geom_line()
        + labs(title="Plotnine Example", x="X-axis", y="Y-axis")
        + theme_minimal())
print(plot.show())
```
